```python
import csv
import random

lines = []
angles = []
with open('./linux_sim/data2/driving_log.csv') as csvfile:
    reader = csv.reader(csvfile)
    for line in reader:
        lines.append(line)
        angles.append(float(line[3]))
```

```python
import matplotlib.pyplot as plt
# Visualizations will be shown in the notebook.
%matplotlib inline

fig, ax_angles = plt.subplots(ncols=1, figsize=(10, 10))

# Steering Angles
ax_angles.hist(angles)
ax_angles.set_xlabel('Steering Angle')
ax_angles.set_ylabel('Test Data Count')
ax_angles.set_title('Distribution of Steering Angles')

# Plot
plt.show()
```

```python
import numpy as np
import random
import matplotlib.image as mpimg
# Example images are from center only.
current_path = './linux_sim/data2/IMG/'
num_ex_images = 2
images = []
cropped_images = []
for i in range(num_ex_images):
    idx = random.randint(0, len(lines))
    img = mpimg.imread(current_path + lines[idx][0].split('/')[-1])
    images.append(img)
    cropped_images.append(img[75:135, :,:])

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16, 16))
# Original
axes[0][0].imshow(images[0])
axes[0][0].set_title('Original')
axes[1][0].imshow(images[1])
axes[1][0].set_title('Original')
# Cropped
axes[0][1].set_title('Cropped')
axes[0][1].imshow(cropped_images[0])
axes[1][1].set_title('Cropped')
axes[1][1].imshow(cropped_images[1])
plt.show()
```
